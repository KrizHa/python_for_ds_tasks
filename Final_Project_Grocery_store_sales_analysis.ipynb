**12. Так як найпопулярнійший товар Ashai Lager Beer  вікової групи яка має найбільше покупок проаналізуємо чи є статистично значуща різниця в середніх чеках для жінок і чоловіків, які купують Ashai Lager Beer. Використаємо статистий тест z-test. Рівень статистичної значущості alpha = 0.05.**

```python
from statsmodels.stats.weightstats import ztest

# Фільтрація даних для товару "Ashai Lager Beer"
filtered_popular_category = df[df['Items'].isin(['Ashai Lager Beer'])]

# Використання z-тесту для порівняння середніх значень суми рахунку між статтями
male_bill_amounts = filtered_popular_category[filtered_popular_category['Gender'] == 'Male']['Gross Bill Amount']
female_bill_amounts = filtered_popular_category[filtered_popular_category['Gender'] == 'Female']['Gross Bill Amount']

# Проведення z-тесту
z_stat, p_value = ztest(male_bill_amounts, female_bill_amounts, value = 0)

# Виведення результатів
print(f"Z-statistic: {z_stat}")
print(f"P-value: {p_value}")

# Нульова гіпотеза - Різниця між середніми значеннями для чоловіків і жінок є значущою
# Альтернативна гіпотеза - Різниця між середніми значеннями для чоловіків і жінок не є значущою

# Перевірка на значущість
alpha = 0.05
if p_value < alpha:
    print("Різниця між середніми значеннями є значущою")
else:
    print("Різниця між середніми значеннями не є значущою")
```

```python
# Обчислення середнього значення загальної суми рахунку за статтю
average_bill_by_gender = filtered_popular_category.groupby('Gender')['Gross Bill Amount'].mean().reset_index()
average_bill_by_gender
```

*Висновок: Чоловікі і жінки мають майже однаковий середній чек при покупці товару Ashai Lager Beer. Але в загальному кількість жінок значно перевищує кількість покупців чоловічої статі. Варто сконцентруватися на залучені більшої кількості чоловіків до покупок в мережі що може сбільшити прибутковість магазинів*
